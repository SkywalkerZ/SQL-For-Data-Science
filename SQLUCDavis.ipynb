```python
from pandasql import sqldf
pysqldf= lambda q: sqldf(q, globals())
```

```python
sport_cleaned=sport_raw.dropna()
```

```python
pysqldf("SELECT * FROM sport_cleaned LIMIT 5")
```

```python
sport_cleaned.columns[[7,8,11,13]]
```

```python
sport_cleaned=sport_cleaned.drop(sport_cleaned.columns[[7,8,11,13]],axis=1)
```

```python
sport_cleaned.head(5)
```

```python
#Checking for any skew values that could affect analysis
pysqldf("SELECT MAX(Age),MIN(Age),MAX(Height),MIN(Height),MAX(Weight),MIN(Weight), MIN(Year),MAX(Year) FROM sport_cleaned")
```

```python
#Checking if there are any more medal types than then 3 main: gold, silver, bronze
pysqldf("SELECT DISTINCT(Medal) FROM sport_cleaned")
```

```python
#Checking if there are any more sex types than then 2 main: Male, Female
pysqldf("SELECT DISTINCT(Sex) FROM sport_cleaned")
```

```python
#Found Team name not corresponding to right country
pysqldf("SELECT * FROM sport_cleaned WHERE Team LIKE '%England%'")
```

```python
#Fixing the Team name to match the country name
for index in sport_cleaned.index:
    if sport_cleaned.loc[index,'Team']=='England-1':
        sport_cleaned.loc[index,'Team'] = 'England'
```

```python
pysqldf("SELECT * FROM sport_cleaned WHERE Team LIKE '%England%'")
```

```python
#Checking the medal distribution between the sex's from cleaned df
pysqldf("SELECT Sex, COUNT(Medal)AS Medals_from_cleaned_df FROM sport_cleaned GROUP BY Sex")
```

```python
#Checking the medal distribution between the sex's from raw df
pysqldf("SELECT Sex, COUNT(Medal) AS Medals_from_raw_df FROM sport_raw WHERE Medal IS NOT NULL GROUP BY Sex")
```

```python
#Checking the medal distribution between the Age 11-20,21-30,31-40,41-50,51-60,61-70 regardless of sex
pysqldf("SELECT CASE \
         WHEN Age>=11 AND Age<=20 THEN 'Age Between 11 to 20' \
         WHEN Age>=21 AND Age<=30 THEN 'Age Between 21 to 30' \
         WHEN Age>=31 AND Age<=40 THEN 'Age Between 31 to 40' \
         WHEN Age>=41 AND Age<=50 THEN 'Age Between 41 to 50' \
         WHEN Age>=51 AND Age<=60 THEN 'Age Between 51 to 60' \
         ELSE 'Age Between 61-70' \
         END AS AgeGroup, \
        COUNT(Medal) AS Medals_from_cleaned_df FROM sport_cleaned GROUP BY AgeGroup")
```

```python
#Checking the medal distribution between the Age 11-20,21-30,31-40,41-50,51-60,61-70 regardless of sex
pysqldf("SELECT CASE \
         WHEN Age>=11 AND Age<=20 THEN 'Age Between 11 to 20' \
         WHEN Age>=21 AND Age<=30 THEN 'Age Between 21 to 30' \
         WHEN Age>=31 AND Age<=40 THEN 'Age Between 31 to 40' \
         WHEN Age>=41 AND Age<=50 THEN 'Age Between 41 to 50' \
         WHEN Age>=51 AND Age<=60 THEN 'Age Between 51 to 60' \
         ELSE 'Age Between 61-70' \
         END AS AgeGroup, \
        COUNT(Medal) AS Medals_from_raw_df FROM sport_raw WHERE Medal IS NOT NULL GROUP BY AgeGroup")
```

```python
#Checking the medal distribution between the Height 131-140,141-150,151-160,161-170,171-180,181-190,191-200,201-210,
#211-220, 221-230 regardless of sex
pysqldf("SELECT CASE \
         WHEN Height>=131 AND Height<=140 THEN 'Height Between 131 to 140' \
         WHEN Height>=141 AND Height<=150 THEN 'Height Between 141 to 150' \
         WHEN Height>=151 AND Height<=160 THEN 'Height Between 151 to 160' \
         WHEN Height>=161 AND Height<=170 THEN 'Height Between 161 to 170' \
         WHEN Height>=171 AND Height<=180 THEN 'Height Between 171 to 180' \
         WHEN Height>=181 AND Height<=190 THEN 'Height Between 181 to 190' \
         WHEN Height>=191 AND Height<=200 THEN 'Height Between 191 to 200' \
         WHEN Height>=201 AND Height<=210 THEN 'Height Between 201 to 210' \
         WHEN Height>=211 AND Height<=220 THEN 'Height Between 211 to 220' \
         ELSE 'Height Between 221-230' \
         END AS HeightGroup, \
        COUNT(Medal) AS Medals_from_cleaned_df FROM sport_cleaned GROUP BY HeightGroup ORDER BY Medals_from_cleaned_df DESC ")
```

```python
pysqldf("SELECT CASE \
         WHEN Weight>=20 AND Weight<=40 THEN 'Weight Between 20 to 40' \
         WHEN Weight>=41 AND Weight<=60 THEN 'Weight Between 41 to 60' \
         WHEN Weight>=61 AND Weight<=80 THEN 'Weight Between 61 to 80' \
         WHEN Weight>=81 AND Weight<=100 THEN 'Weight Between 81 to 100' \
         WHEN Weight>=101 AND Weight<=120 THEN 'Weight Between 101 to 120' \
         WHEN Weight>=121 AND Weight<=140 THEN 'Weight Between 121 to 140' \
         WHEN Weight>=141 AND Weight<=160 THEN 'Weight Between 141 to 160' \
         WHEN Weight>=161 AND Weight<=180 THEN 'Weight Between 161 to 180' \
         ELSE 'Weight Between 181-200' \
         END AS WeightGroup, \
        COUNT(Medal) AS Medals_from_cleaned_df FROM sport_cleaned GROUP BY WeightGroup ORDER BY Medals_from_cleaned_df DESC")
```

```python
#Check which countries have more medals (sign of gifted athletes)
pysqldf("SELECT Team, COUNT(Medal) AS COUNT FROM sport_cleaned GROUP BY Team ORDER BY COUNT DESC LIMIT 10")
```

```python
#Check which countries have more medals (sign of gifted athletes)
pysqldf("SELECT Sport, COUNT(Medal) AS Medal_COUNT, AVG(Height), AVG(Weight), AVG(Age) FROM sport_cleaned GROUP BY Sport ORDER BY Medal_COUNT DESC")
```
